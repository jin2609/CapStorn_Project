```python
#의미없는 변수 삭제
del data_employee['x']
del data_employee['x.1']
del data_employee['x.2']
del data_employee['x.3']
del data_employee['x.4']
del data_employee['x.5']
del data_employee['x.6']
del data_employee['x.7']
del data_employee['x.8']
for col in data_employee.columns:
    print(col,":",len(data_employee[col].unique()))
    if(len(data_employee[col].unique()) == 1):
        del data_employee[col]
```

```python
#데이터 전처리1 _ Employee관련 key : 사번
EmployeeData = pd.DataFrame()
EmployeeData['EmployeeNum'] = data_employee['사번']
EmployeeData.head()
```

```python
#유도변수 생성 생년월일 -> 나이 
year = []
for date in data_employee['생년월일'] :
    year.append(2019-(date.year)+1)
EmployeeData['Year'] = year
EmployeeData.head()
```

```python
#성별 처리 남자 : 0, 여자 : 1
sex = []
for Sex in data_employee['남'].fillna(1):
    if Sex == 1 :
        sex.append(1)
    else :
        sex.append(0)
EmployeeData['Female'] = sex
EmployeeData.head()
```

```python
#학습 처리 중졸이상 : 0 고졸이상 : 1 대졸이상 : 2
edu = []

for row in data_employee.index :
    if(data_employee['중졸이상'][row] == 'O'):
        edu.append(0)
    if(data_employee['고졸이상'][row] == 'O'):
        edu.append(1)
    if(data_employee['대졸이상'][row] == 'O'):
        edu.append(2)
EmployeeData['Edu'] = edu
EmployeeData.head()
```

```python
#작업공정 서브 = SU 가공 = GC 조립 = LR 검사 =QC 포장 = DC 관리 = MM
part = []

for row in data_employee.index :
    flag = 0
    if(data_employee['서브'][row] == 'O'):
        part.append('SU')
        flag = 1
    if(data_employee['가공'][row] == 'O'):
        part.append('GC')
        flag = 1
    if(data_employee['조립'][row] == 'O'):
        part.append('LR')
        flag = 1
    if(data_employee['검사'][row] == 'O'):
        part.append('QC')
        flag = 1
    if(data_employee['포장'][row] == 'O'):
        part.append('DC')
        flag = 1
    if(flag != 1):
        part.append('MM')
EmployeeData['Part'] = part
EmployeeData.head()
```

```python
#Skill 숙련도 C : 0 B : 1 A : 2 S : 3 관리자 : 4
skill = []

for row in data_employee.index :
    flag = 0
    if(data_employee['C'][row] == 'O'):
        skill.append(0)
        flag = 1
    if(data_employee['B'][row] == 'O'):
        skill.append(1)
        flag = 1
    if(data_employee['A'][row] == 'O'):
        skill.append(2)
        flag = 1
    if(data_employee['S'][row] == 'O'):
        skill.append(3)
        flag = 1
    if(flag != 1):
        skill.append(4)
EmployeeData['Skill'] = skill
EmployeeData.head()
```

```python
#hand 주로 사용하는 손 왼손 : 0 오른손 : 1 양손 : 2
hand = []

for row in data_employee.index :
    if(data_employee['왼손잡이'][row] == 'O'):
        hand.append(0)
    if(data_employee['오른손잡이'][row] == 'O'):
        hand.append(1)
    if(data_employee['양손잡이'][row] == 'O'):
        hand.append(2)
EmployeeData['Hand'] = hand
EmployeeData.head()
```

```python
#Service  용역직 : 1 직접 고용: 0
ser = []

for service in data_employee['용역직'].fillna(1):
    if service == 1 :
        ser.append(0)
    else :
        ser.append(1)
EmployeeData['Service'] = ser
EmployeeData.head()
```

```python
#Home 본적지 지역주민 : 1 타 지역주민 : 0
home = []

for hom in data_employee['지역주민'].fillna(1):
    if hom == 1 :
        home.append(0)
    else :
        home.append(1)
EmployeeData['Home'] = home
EmployeeData.head()
```

```python
#NumberofYear 근속년수 1개월 미만 : 0, 1개월 이상 : 1, 3개월 이상 : 2, 6개월 이상 : 3, 1년 이상 : 4, 3년 이상 : 5
noy = []
for row in data_employee.index :
    flag = 0
    if(data_employee['근속년수_1개월미만'][row] == 'O'):
        noy.append(0)
    if(data_employee['근속년수_1개월이상'][row] == 'O'):
        noy.append(1)
    if(data_employee['근속년수_3개월이상'][row] == 'O'):
        noy.append(2)
    if(data_employee['근속년수_6개월이상'][row] == 'O'):
        noy.append(3)
    if(data_employee['근속년수_1년이상'][row] == 'O'):
        noy.append(4)
    if(data_employee['근속년수_3년이상'][row] == 'O'):
        noy.append(5)
EmployeeData['NumberofYear'] = noy
EmployeeData.head()
```

```python
#Smoke 담배 태우면 : 1 안태우면 0
smo = []
for sm in data_employee['흡연'].fillna(1):
    if sm == 1 :
        smo.append(0)
    else :
        smo.append(1)
EmployeeData['Smoke'] = smo
EmployeeData.head() 
```

```python
#drink 음주자 : 1 비음주자 : 0
drink = []
for dr in data_employee['음주O'].fillna(1):
    if dr == 1 :
        drink.append(0)
    else :
        drink.append(1)
EmployeeData['Drink'] = drink
EmployeeData.head()     
```

```python
#married 기혼 : 1 미혼 : 0
mar = []
for ma in data_employee['기혼'].fillna(1):
    if ma == 1 :
        mar.append(0)
    else :
        mar.append(1)
EmployeeData['Married'] = mar
EmployeeData.head()
```

```python
#Chiled 자녀O : 1 없으면 : 0
chiled = []
for chi in data_employee['자녀 O'].fillna(1):
    if chi == 1:
        chiled.append(0)
    else :
        chiled.append(1)
EmployeeData['Chiled'] = chiled
EmployeeData.head()
```

```python
#Pregnancy 임신O : 1 임신x : 0
pre = []
for pr in data_employee['임신 O'].fillna(1):
    if(pr == 1):
        pre.append(0)
    else :
        pre.append(1)
EmployeeData['Pregnancy'] = pre
EmployeeData.head()
```

```python
#통근 방법
#Co_Car, Co_bicycle, ,Co_bus, Co_Foot, Co_Other
car = []
bicycle = []
bus = []
foot = []
other = []

for row in data_employee.index :
    if(data_employee['자차'][row] == 'O'):
        car.append(1)
        bicycle.append(0)
        bus.append(0)
        foot.append(0)
        other.append(0)
    if(data_employee['자전거'][row] == 'O'):
        car.append(0)
        bicycle.append(1)
        bus.append(0)
        foot.append(0)
        other.append(0)
    if(data_employee['통근버스'][row] == 'O'):
        car.append(0)
        bicycle.append(0)
        bus.append(1)
        foot.append(0)
        other.append(0)
    if(data_employee['도보'][row] == 'O'):
        car.append(0)
        bicycle.append(0)
        bus.append(0)
        foot.append(1)
        other.append(0)
    if(data_employee['기타'][row] == 'O'):
        car.append(0)
        bicycle.append(0)
        bus.append(0)
        foot.append(0)
        other.append(1)
EmployeeData['Co_Car'] = car
EmployeeData['Co_bicycle'] = bicycle
EmployeeData['Co_bus'] = bus
EmployeeData['Co_onFoot'] = foot
EmployeeData['Co_Other'] = other
EmployeeData.head()
```

```python
#CommutingTime=> 통근 시간 30분 미만 : 0, 30분 이상 : 1, 1시간 이상 : 2
time = []
for row in data_employee.index :
    if(data_employee['통근시간_30분미만'][row] == 'O'):
        time.append(0)
    if(data_employee['통근시간_30분이상'][row] == 'O'):
        time.append(1)
    if(data_employee['통근시간_1시간이상'][row] == 'O'):
        time.append(2)
EmployeeData['CommutingTime'] = time
EmployeeData.tail()
```

```python
#시력 처리 
vision = []
for vis in data_employee['시력']:
    visi = vis.split('/')
    vision.append(((int)(visi[0]) + (int)(visi[1]))/ 2)
EmployeeData['Vision'] = vision
EmployeeData.head()
```

```python
#민족,급여,키,몸무게,3월 근무시간(주)-잔업/특근포함, 3월 특근 횟수(월)
EmployeeData['EthnicGroup'] = data_employee['민족']
EmployeeData['Payment'] = data_employee['급여']
EmployeeData['Height'] = data_employee['키']
EmployeeData['Weight'] = data_employee['몸무게']
EmployeeData['DutyHours'] = data_employee['3월 근무시간 ㈜ \n-잔업/특근 포함']
EmployeeData['OverTimeWork'] = data_employee['3월 특근 횟수(월)']
EmployeeData.head()
```

```python
EmployeeData.to_csv('./data/Employee.csv',index=False)
```

## 2. raw 데이터 정제 : 제품 데이터
### -> 각 변수에 대한 class 지정

```python
data_product = pd.read_excel('./data/product_rawdata.xlsx')
data_product.head()
```

```python
badFlag = True
df_product = pd.DataFrame(columns=['Bad','badCd','ptGroupNm','qabDscr','work_time','가공','EmployeeNum'])
size = 0;
index = 0;
for i in range(len(data_product)):
    if(data_product['작업시간'].fillna(-1)[i] == -1):
        badFlag = False
        continue
    if(data_product['가공/조립구분'].fillna(-1)[i] != -1):
        badFlag = True
        index = i
    if(badFlag):
        df_product.loc[size] = [1,data_product['badCd'][index],data_product['ptGroupNm'][index],data_product['qabDscr'][index],data_product['작업시간'][i],data_product['가공/조립구분'][index],data_product['DB사번'][i]]
    else:
        df_product.loc[size] = [0,data_product['badCd'][index],data_product['ptGroupNm'][index],data_product['qabDscr'][index],data_product['작업시간'][i],data_product['가공/조립구분'][index],data_product['DB사번'][i]]
    size = size + 1
df_product.head()
```

```python
df_product.to_csv('./data/Product.csv',index=False, encoding='utf-8-sig')
```

## 3. merge를 활용한 데이터 결합

```python
employee = pd.read_csv('./data/Employee.csv')
employee.head()
```

```python
product = pd.read_csv('./data/Product.csv')
product.head()
```

```python
## 필요없는 칼럼 삭제
data = pd.merge(product,employee,on ='EmployeeNum', how = 'left')
data = data.dropna()
data = data[data['가공'] == '가공']
del data['가공']
del data['work_time']
del data['EmployeeNum']
del data['Service']
del data['Co_bus']
del data['Co_Other']
del data['EthnicGroup']
data = data[data['Part'] == 'GC']
del data['Vision']
del data['Co_Car']
del data['Co_bicycle']
del data['Co_onFoot']
del data['Drink']
del data['Smoke']
del data['Part']
del data['Home']
data.shape
```

```python
data.to_csv('./data/process_data.csv',index=False, encoding='utf-8-sig')
```
